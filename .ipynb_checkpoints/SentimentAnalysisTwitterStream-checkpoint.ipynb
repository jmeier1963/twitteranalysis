# Analysis Twitter Stream 

```python
import pandas as pd
import os

q = "#Suffizienz"
since = "2010-01-01"
until = "2020-6-30"

tweets2 = pd.read_pickle(q + "_tweets_" + since + "_to_" + until +".pkl")
```

```python
tweets2.info()
```

# Text Processing with the Natural Language Toolkit

```python
import nltk
from nltk.corpus import stopwords
from nltk import FreqDist
```

```python
stop_eng = stopwords.words('english')
stop_ger = stopwords.words('german')
customstopwords = ['Einwanderung', 'einwanderung', 'wir', 'erst', 'sollten', 'az', 'su', 'gibt', 'wir', 'neben', 'twitter', 'erste', 'vielen', 'aw' ]
```

```python
rawtext = " ".join(tweets2["text"]).lower()
tokens = []
sentences = []
for txt in rawtext.split():
    tokens.extend([t.lower().strip(":,.!?") for t in txt.split()])

hashtags = [w for w in tokens if w.startswith('#')]
mentions = [w for w in tokens if w.startswith('@')]
links = [w for w in tokens if w.startswith('http') or w.startswith('www')]
filtered_tokens = [w for w in tokens \
                   if not w in stop_eng \
                   and not w in stop_ger \
                   and not w in customstopwords \
                   and w.isalpha() \
                   and not len(w)<3 \
                   and not w in hashtags \
                   and not w in links \
                   and not w in mentions]
```

## Concordance

```python
s = " ".join(tweets2["text"])
s_tokens = nltk.word_tokenize(s)
```

```python
rawtweettext = nltk.Text(s_tokens)
rawtweettext.concordance("Kapitalismus")
```

```python
rawtweettext.similar('Kapitalismus')
```

## Collocations

In corpus linguistics, a collocation is a sequence of words or terms that co-occur more often than would be expected by chance.

```python
tweettext = nltk.Text(filtered_tokens)
tweettext.collocations()
```

## Search for Words

```python
fdist = nltk.FreqDist([w.lower() for w in tweettext])
modals = ['kapitalismus', 'verzicht', 'teilhabe', 'verbot', 'regulierung']
for m in modals:
    print(m + ':', fdist[m])
```

## Dispersion Plot

Determine the location of a word in the text: how many words from the beginning it appears. This positional information can be displayed using a dispersion plot.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,5))
rawtweettext.dispersion_plot(modals)
```

## Sentiment Analysis

### Sentiment analysis with Textblob-de

#### Example usage

https://textblob-de.readthedocs.io/en/latest/readme.html

https://textblob.readthedocs.io/en/dev/

```python
from textblob_de import TextBlobDE as TextBlob
```

```python
text1 = "Das ist alles wunderschön. Ich freue mich."
text2 = 'Heute ist der 3. Mai 2014 und Dr. Meier feiert seinen 43. Geburtstag. Ich muss unbedingt daran denken, Mehl, usw. für einen Kuchen einzukaufen. Aber leider habe ich nur noch EUR 3,50 in meiner Brieftasche.'
text3 = 'Es ist so schlimm und so furchtbar. Es war alles so traurig und grässlich.'
blob = TextBlob(text1)
print(blob.sentences)
print(blob.tokens)
print(blob.tags)
print(blob.noun_phrases)
print(blob.sentiment)

blob2 = TextBlob(text2)
print(blob2.sentiment)

blob3 = TextBlob(text3)
print(blob3.sentiment)
```

```python
print(blob3.sentiment.polarity)
```

### Application to Tweets

```python
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

tweets2["polarity"]=tweets2["text"].apply(get_sentiment)
```

```python
def get_sentiment_category(p):
    if p > 0.4: 
        return "positive" 
    elif p < -0.4: 
        return "negative" 
    else: 
        return "neutral"

tweets2["sentiment"] = tweets2["polarity"].apply(get_sentiment_category)
```

```python
tweets2["sentiment"].hist(bins=30)
```

```python
tweets2.set_index('date', inplace=True)
tweets2.index = tweets2.index.tz_localize('GMT').tz_convert('CET')
tweets2.index.name = 'zeit'
tweets2.index
```

```python
plt.figure(figsize=(16,12))
s = tweets2["polarity"].resample('M').mean().plot(label='mood of the Twitter crowd', color='g', alpha=0.8)
plt.annotate('positive', xy=(0.4, 0.8),  xycoords='axes fraction',
            xytext=(0.5, 0.8), textcoords='axes fraction', size=16,
            horizontalalignment='center', verticalalignment='center')
plt.annotate('negative', xy=(0.5, 0.8),  xycoords='axes fraction',
            xytext=(0.5, 0.2), textcoords='axes fraction', size=16,
            horizontalalignment='center', verticalalignment='center')
plt.legend(loc=4)
plt.title('Mood of the Twitter Crowd estimated with sum of polarity of tweets of the week')
plt.savefig('%s-mood-crowd.png' % q, bbox_inches='tight', dpi=300)
```
